### How to generate Insights from a Dataset

Differenciate between: 
    Values and Labels
    
    Values: Sales, Quantity, Discount, Profit
    Labels: Ship Mode, Segements, Country, States, City, Region, Category, Sub-Category, Product Name, Order Date
    Aggregates: .sum(), .max(), .min(), .count(), .mean()
    
Bring them all together to find Insights:
    Total number of Profit made in each States.
    Total Sales made in each region.
    Average Disount given to Segements
    Largest amount of Profit in a Category.
    Total Quantity Sold in each Region
    Total Quantity Sold in State
    Most prefered shipping Method

### How to Filter



```python
dataset['City'].count()
```

```python
# How to get a Column Data
dataset['Product Name']
```

```python
# Filtering Orders in East Region alone
dataset[(dataset['Region'] == 'East')]

```

```python
# Filtering Orders in 'East Region' and 'Category is Furniture' and 'Ship Mode is Second Class'

dataset[(dataset['Region'] == 'East') & 
        (dataset['Category'] == 'Furniture') & 
        (dataset['Ship Mode'] == 'Second Class')]
```

```python
# Filtering with Alternatives

dataset[(dataset['Region'].any() == 'East') or (dataset['Region'] == 'South')] # & (dataset['Ship Mode'] == 'First Class')]
```

### How to sort in Ascending or Descending Order

```python
dataset.sort_values(by='Product Name') # A - Z

dataset.sort_values(by='City') # A - Z

dataset.sort_values(by='City', ascending=False) # Z - A
```

```python
dataset.sort_values(by='Sales', ascending = False) # Largest - Smallest

dataset.sort_values(by='Sales') # Smallest - Largest
```
